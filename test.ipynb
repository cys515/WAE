```python

 
import matplotlib.pyplot as plt
import dill
import torch
from torch import nn
from matplotlib.colorbar import ColorbarBase as cb
device = torch.device("cuda:1" if torch.cuda.is_available() else "cpu")
with open('./Datasets/L160.dill', 'rb') as f1:
    data = dill.load(f1)
    L1 = data[2][0:20]
    L1Y = data[3][0:20]
with open('./Datasets/P160.dill', 'rb') as f1:
    data = dill.load(f1)
    P1 = data[2][0:20]
    P1Y = data[3][0:20]
with open('./Datasets/P560.dill', 'rb') as f1:
    data = dill.load(f1)
    P4 = data[2][0:20]
    P4Y = data[3][0:20]
with open('./Datasets/pufa60.dill', 'rb') as f1:
    data = dill.load(f1)
    PF = data[2][0:20]
    PFY = data[3][0:20]

		
#load black box model 
model_name='./Models/L1_LSTM60'
# model = keras.models.load_model(model_name)
model1 = torch.load(model_name).to(device)   
model_name='./Models/P1_LSTM60'
# model = keras.models.load_model(model_name)
model2 = torch.load(model_name).to(device)   
model_name='./Models/P4_LSTM60'
# model = keras.models.load_model(model_name)
model3 = torch.load(model_name).to(device)   
model_name='./Models/pufa_LSTM60'
# model = keras.models.load_model(model_name)
model4 = torch.load(model_name).to(device)   

model1.eval()
model2.eval()
model3.eval()
model4.eval()
L1 = torch.from_numpy(L1).float().to(device)
P1 = torch.from_numpy(P1).float().to(device)
P4 = torch.from_numpy(P4).float().to(device)
PF = torch.from_numpy(PF).float().to(device)
L1Y = torch.from_numpy(L1Y).float().to(device)
P1Y = torch.from_numpy(P1Y).float().to(device)
P4Y = torch.from_numpy(P4Y).float().to(device)
PFY = torch.from_numpy(PFY).float().to(device)

prediction1=model1(L1)
prediction2=model2(P1)
prediction3=model3(P4)
prediction4=model4(PF)

cri = nn.MSELoss()
m1 = cri(prediction1,L1Y)
m2 = cri(prediction2,P1Y)
m3 = cri(prediction3,P4Y)
m4 = cri(prediction4,PFY)

print(m1)
print(m2)
print(m3)
print(m4)
```
